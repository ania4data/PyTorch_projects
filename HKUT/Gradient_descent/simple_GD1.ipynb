```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w=0.0
epochs=100
lr=0.001 #learning rate
x_data=[1.0,2.0,3.0]
y_data=[2.0,4.0,6.0]
```

```python
def forward(x):
    return w*x
def loss(x,y):
    y_pred=forward(x)
    return (y-y_pred)**2.0
def gradient(x,y):
    return 2.0*(y-forward(x))*x*-1.0
    

loss_=[]
w_=[]

fig= plt.figure()

axes= fig.add_axes([0,0,1,1])

for epoch in range(epochs):
    
    for i,(x,y) in enumerate(zip(x_data,y_data)): #some thing like stochstic, not whole vector at once
        
        w=w-lr*gradient(x,y)
        
    if (epoch%5==0): print("epoch   %d / %d  : %f   %f" %(epoch,epochs,loss(x,y),w))
    loss_.append(loss(x,y))
    w_.append(w)
    xx=np.arange(0,6,1)
    yy=xx*w
    axes.plot(xx,yy,linewidth=w/1.0,color='blue')
    

plt.show()    
plt.plot(w_,loss_)
plt.show()
    
    
    
    
    
    
    
    
    
    
```
