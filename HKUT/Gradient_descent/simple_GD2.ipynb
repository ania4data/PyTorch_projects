```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1=0.0
w2=0.0
b=0.0
epochs=100
lr=0.001 #learning rate
x_data=[1.0,2.0,3.0]
y_data=[2.0,4.0,6.0]
```

```python
def forward(x):
    return (w2*(x**2.0))+(w1*x)+b
def loss(x,y):
    y_pred=forward(x)
    return (y-y_pred)**2.0
def gradient_w2(x,y):
    return 2.0*(y-forward(x))*(x**2.0)*-1.0
def gradient_w1(x,y):
    return 2.0*(y-forward(x))*x*-1.0  
def gradient_b(x,y):
    return 2.0*(y-forward(x))*-1.0*1.0  

loss_=[]
w_=[]

fig= plt.figure()

axes= fig.add_axes([0,0,1,1])

for epoch in range(epochs):
    
    for i,(x,y) in enumerate(zip(x_data,y_data)): #some thing like stochstic, not whole vector at once
        
        w2=w2-lr*gradient_w2(x,y)
        w1=w1-lr*gradient_w1(x,y)
        b=b-lr*gradient_b(x,y)
        
        
    if (epoch%5==0): print("epoch   %d / %d  : %f   %f %f %f" %(epoch,epochs,loss(x,y),w1,w2,b))
    loss_.append(loss(x,y))
    w_mag=np.sqrt(w1**2.0+w2**2.0+b**2.0)
    w_.append(w_mag)
    xx=np.arange(0,6,1)
    yy=(w2*(xx**2.0))+(w1*xx)+b
    axes.plot(xx,yy,linewidth=w_mag/1.0,color='blue')
    

plt.show()    
plt.plot(w_,loss_)
plt.show()

plt.plot(np.arange(0,epochs,1),w_)   #need to detach gradient for graph breakdown otherwise won't do numpy in the middle of graph (good for end of run)
plt.show()    
plt.plot(np.arange(0,epochs,1),loss_)
plt.show()
    

    
```
